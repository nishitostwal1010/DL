```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('dog.jpg', target_size=(228,228))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=30, #Range by which it can be rotated in +ve or -ve direction
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    #vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='nearest' #If we are shifting image and no pixels are present then fill them using nearby pixels
    # If we use constant fill mode then the pixels which are not present will be shown as black space
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,228,228,3)
```

```python
!mkdir 'aug'
```

```python
i = 0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='aug'):
    i=i+1

    if i == 5:
        break
```
